### Read the dataset for churn

```python
df = pd.read_csv("/content/drive/MyDrive/smart_systems/Coursework/data/script/new_input.csv")
print('new data size', df.shape)
```

```python
# Define target and features columns
y_column = ['Quality']
X_columns = ['Size','Weight', 'Sweetness', 'Crunchiness', 'Juiciness',
                 'Ripeness', 'Acidity']

X = df[X_columns]
y = df[y_column]
try:
    rf = pickle.load(open('/content/drive/MyDrive/smart_systems/Coursework/data/script/finalized_model.sav', 'rb'))
    y_pred = rf.predict(X)
    print('test set metrics: ', metrics.classification_report(y, y_pred))
    df['Quality_pred'] = rf.predict(X)

    # Save the prediction results to a CSV file
    df.to_csv('/content/drive/MyDrive/smart_systems/Coursework/data/script/prediction_results.csv', index=False)
except EOFError as e:
    print("Error loading the model:", e)
```
