### Read the dataset for churn

```python
df = pd.read_csv("/content/drive/MyDrive/smart_systems/Coursework/data/script/train.csv")
```

```python
# Define target and features columns
y_column = ['Quality']
X_columns = ['Size','Weight', 'Sweetness', 'Crunchiness', 'Juiciness',
                 'Ripeness', 'Acidity']
X = df[X_columns]
y = df[y_column]

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

best_params = {'C': 10, 'cache_size': 200, 'class_weight': None, 'coef0': 0.0, 'degree': 2, 'gamma': 'scale', 'kernel': 'rbf', 'probability': False, 'shrinking': True, 'tol': 0.001}
svc_model = SVC(**best_params)

# Fit the model on the training data
svc_model.fit(X_train, y_train)

# Save the trained model
filename = '/content/drive/MyDrive/smart_systems/Coursework/data/script/finalized_model.sav'
pickle.dump(svc_model, open(filename, 'wb'))
```
